Set up Data

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import numpy as np

iris = load_iris()
X = iris.data[:, [2,3]]
y = iris.target
print(f"Class labels: {np.unique(y)}")

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1, stratify=y
    )

sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
from models import LogisticRegressionGD
from utils import plot_decision_regions
import matplotlib.pyplot as plt

X_train_01_subset = X_train_std[(y_train == 0) | (y_train == 1)]
y_train_01_subset = y_train[(y_train == 0) | (y_train == 1)]
lr = LogisticRegressionGD(eta=0.3, n_iter=1000, random_state=1)

lr.fit(X_train_01_subset, y_train_01_subset)
plot_decision_regions(X=X_train_01_subset, 
                      y=y_train_01_subset, 
                      classifier=lr)
plt.xlabel('Petal length [standardized]')
plt.ylabel('Petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

Using Scikit-Learn

```python
from sklearn.linear_model import LogisticRegression
lr_sk = LogisticRegression(C=100.0, solver='lbfgs')
lr_sk.fit(X_train_std, y_train)
X_combined = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))
plot_decision_regions(X=X_combined, y=y_combined, classifier=lr_sk, 
                      mark_split=True, test_idx=range(len(y_train), len(y_combined)))
plt.xlabel('Petal length [standardized]')
plt.ylabel('Petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

Regularization

```python
weights,params = [], []
for c in np.arange(-5,5):
    lr = LogisticRegression(C=10.**c, solver='lbfgs')
    lr.fit(X_train_std, y_train)
    weights.append(lr.coef_[1])
    params.append(10.**c)

weights = np.array(weights)
plt.plot(params, weights[:, 0], label = 'Petal Length')
plt.plot(params, weights[:, 1], linestyle = '--', label = 'Petal Width')
plt.ylabel('Weight coefficient')
plt.xlabel('C')
plt.legend(loc='upper left')
plt.xscale('log')
plt.show()
```
