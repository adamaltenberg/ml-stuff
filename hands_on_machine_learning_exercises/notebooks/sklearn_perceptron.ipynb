```python
from sklearn.datasets import load_iris
import numpy as np

iris = load_iris()
X = iris.data[:, [2,3]]
y = iris.target
print(f"Class labels: {np.unique(y)}")

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1, stratify=y
    )
print(f"""\nLabel counts in 
y: {np.bincount(y)}
y_train: {np.bincount(y_train)}
y_test: {np.bincount(y_test)}""")
```

Standardizing Features

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

Perceptron Model

```python
from sklearn.linear_model import Perceptron
ppn = Perceptron(eta0=0.1, random_state=1)
ppn.fit(X_train_std, y_train)
y_pred = ppn.predict(X_test_std)
```

Testing Accuracy

```python
from sklearn.metrics import accuracy_score
from utils import plot_decision_regions
import matplotlib.pyplot as plt

print(f"Misclassified examples: {(y_test != y_pred).sum():d}")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.3f} (native method {ppn.score(X_test_std,y_test):.3f})")
X_combined = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))
plot_decision_regions(ppn, X_combined, y_combined)
plt.xlabel('Sepal length (standardized)')
plt.ylabel('Petal length (standardized)')
plt.show()
```
